```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("Netflix Dataset.csv")
data.head()
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
data['Release_Date'] = pd.to_datetime(data['Release_Date'], format='%B %d, %Y', errors='coerce')
data['Release_Year'] = data['Release_Date'].dt.year
```

```python
data.head()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(y='Category', data=data, palette='coolwarm')
plt.title("Distribution of Content Type")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='Release_Year', data=data, palette='magma')
plt.title("Number of Titles Released per Year")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(y='Rating', data=data, order=data['Rating'].value_counts().index, palette='viridis')
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Release_Year', hue='Category', data=data, palette='Set2')
plt.title("Movies vs TV Shows by Release Year")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_countries = data['Country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='cool')
plt.title("Top 10 Countries with Most Titles")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Rating', hue='Category', data=data, palette='Set2')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Split multiple countries into separate rows
data['Country'] = data['Country'].astype(str)
data['Country'] = data['Country'].str.replace(' ,', ',')  # clean stray commas
data['Country'] = data['Country'].str.replace(', ', ',')  # normalize commas
data['Country'] = data['Country'].str.split(',')

# Expand rows for each country
df_exploded = data.explode('Country')

# Clean whitespace
df_exploded['Country'] = df_exploded['Country'].str.strip()
```

```python
import plotly.express as px
pivot = df_exploded.pivot_table(index='Country', columns='Category', values='Show_Id', aggfunc='count').fillna(0)
pivot['Total_Titles'] = pivot.sum(axis=1)
map_df = pivot.reset_index()
bins = [0, 5, 10, 20, 50, 100, 200, np.inf]
labels = ['1–5', '6–10', '11–20', '21–50', '51–100', '101–200', '200+']
map_df['Title_Range'] = pd.cut(map_df['Total_Titles'], bins=bins, labels=labels, include_lowest=True)
fig = px.choropleth(
    map_df,
    locations="Country",
    locationmode="country names",
    color="Title_Range",
    hover_name="Country",
    color_discrete_sequence=px.colors.sequential.YlOrRd,
    title="Netflix Titles by Country (Binned)"
)
fig.update_geos(showframe=False, showcoastlines=False)
fig.update_layout(margin={"r":0,"t":50,"l":0,"b":0})
fig.show()
```

```python
plt.figure(figsize=(10,6))
rating_category = pd.crosstab(data['Rating'], data['Category'])
sns.heatmap(rating_category, annot=True, fmt='d', cmap='YlOrRd')
plt.title("Heatmap: Rating vs Category")
plt.show()
```

```python
df_genre = data.copy()
df_genre['Type'] = df_genre['Type'].astype(str)
df_genre['Genre_List'] = df_genre['Type'].str.split(', ')
df_genre = df_genre.explode('Genre_List')

# Clean genre names
df_genre['Genre_List'] = df_genre['Genre_List'].str.strip()

# Top 10 genres overall
top_genres = df_genre['Genre_List'].value_counts().head(10)
print("Top 10 Genres:\n", top_genres)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='magma')
plt.title("Top 10 Most Common Netflix Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
top5_genres = top_genres.index[:5]
genre_trends = (
    df_genre[df_genre['Genre_List'].isin(top5_genres)]
    .groupby(['Release_Year', 'Genre_List'])
    .size()
    .reset_index(name='Count')
)

plt.figure(figsize=(12,6))
sns.lineplot(data=genre_trends, x='Release_Year', y='Count', hue='Genre_List', marker='o')
plt.title("Popularity Trend of Top 5 Genres Over Time")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.legend(title="Genre")
plt.tight_layout()
plt.show()
```

```python
df_country = data.copy()
df_country['Country'] = df_country['Country'].astype(str).str.split(',')
df_country = df_country.explode('Country')
df_country['Country'] = df_country['Country'].str.strip()

# Top 15 contributing countries
top_countries = df_country['Country'].value_counts().head(15)
print("🌎 Top 15 Countries by Content Count:\n", top_countries)
top5_countries = top_countries.index[:5]
country_trends = (
    df_country[df_country['Country'].isin(top5_countries)]
    .groupby(['Release_Year', 'Country'])
    .size()
    .reset_index(name='Count')
)

plt.figure(figsize=(12,6))
sns.lineplot(data=country_trends, x='Release_Year', y='Count', hue='Country', marker='o')
plt.title("Top 5 Countries – Content Production Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.legend(title="Country")
plt.tight_layout()
plt.show()
```
